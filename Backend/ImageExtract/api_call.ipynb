```python
from google import genai
from google.genai import types

from ImageExtract.prompt import (
    AGENT_MESSAGE,
    TYPE_MESSAGE,
)

with open('/home/hoanganh04/Projects/data4life/ImageExtract/data/giay-phep-lai-xe-a1.jpg', 'rb') as f:
    image_bytes = f.read()

client = genai.Client()
response = client.models.generate_content(
model='gemini-2.5-flash',
contents=[
    types.Part.from_bytes(
    data=image_bytes,
    mime_type='image/jpeg',
    ),
    AGENT_MESSAGE + TYPE_MESSAGE
]
)

print(response.text)
```
